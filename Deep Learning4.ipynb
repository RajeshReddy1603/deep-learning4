```python
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from keras.callbacks import ModelCheckpoint
df=pd.read_csv("pima-indians-diabetes.csv",header=None)
df
```

```python
x=df.iloc[:,0:8].values
y=df.iloc[:,8].values
```

```python
model=Sequential()
layers12=Dense(10,input_shape=(8,),activation="relu")
layers3=Dense(20,activation="relu")
layers4=Dense(1,activation="sigmoid")
model.add(layers12)
model.add(layers3)
model.add(layers4)
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
filepath="weights-improvements-{epoch:02d}-{val_accuracy:.2f}.keras"
```

```python
checkpoint=ModelCheckpoint(filepath,monitor="val_accuracy",verbose=0,save_best_only=True,mode="max")
```

```python
callbacks_list=[checkpoint]
```

```python
model.fit(x,y,validation_split=0.33,epochs=100,batch_size=10,callbacks=callbacks_list)
```
